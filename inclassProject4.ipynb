```python
import pandas_datareader.data as web
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
```

```python
start = '2000-01-01'
end = dt.datetime.today().strftime('%Y-%m-%d')

# Unemployment Rate data from FRED
unrate_data = web.DataReader("UNRATE", "fred", start, end)

# I think the sample code say it is inflation?

plt.figure(figsize=(12, 6))
unrate_data.plot(title="US Unemployment Rate (UNRATE)", legend=False)
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
```

```python
# CPI data from FRED
cpi_data = web.DataReader("CPIAUCSL", "fred", start, end)

plt.figure(figsize=(12, 6))
cpi_data.plot(title="US Consumer Price Index (CPIAUCSL)", legend=False)
plt.xlabel('Date')
plt.ylabel('CPI')
```

```python
data = pd.concat([unrate_data, cpi_data], axis=1)
data.columns = ['unrate', 'cpi']
data['cpi_lag'] = data['cpi'].shift()
# data['previous_date'] = data['date'].shift()
print(data.head())
data.describe()
```
