```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller

# Assuming you have downloaded the data and it's stored in CSV files for simplicity
# Replace 'path_to_file.csv' with the actual file paths
unrate = pd.read_csv('C:\\2024\\CRA\\5\\UNRATE.csv')
oil_prices = pd.read_csv('C:\\2024\\CRA\\5\\DCOILBRENTEU.csv')
gdp = pd.read_csv('C:\\2024\\CRA\\5\\GDP.csv')
treasury_rates = pd.read_csv('C:\\2024\\CRA\\5\\T10Y2Y.csv')
volatility_series = pd.read_csv('C:\\2024\\CRA\\5\\VIXCLS.csv')
# Creating a GDP growth variable
gdp['GDP_growth'] = (gdp['GDP'] - gdp['GDP'].shift(1)) / gdp['GDP'].shift(1)

# Assuming you've already prepared your charge-off data, let's merge them
# We'll need a common date column in both datasets to merge on, e.g., 'Quarter'
# This is a simplification; in practice, you'd ensure the dates align correctly
economic_data = pd.DataFrame({
    'DATE': gdp['DATE'],  # Ensure this aligns with your economic data
    'UNRATE': unrate['UNRATE'],
    'DCOILBRENTEU': oil_prices['DCOILBRENTEU'],
    'GDP_Growth': gdp['GDP_growth'],
    'T10Y2Y': treasury_rates['T10Y2Y'],
    'VIXCLS': volatility_series['VIXCLS']  # Replace with your chosen series
}).dropna()  # Dropping missing values after shifting for GDP growth



economic_data['DATE'] = pd.to_datetime(economic_data['DATE'])
economic_data['DATE'] = economic_data['DATE'] - pd.Timedelta(days=1)

economic_data['UNRATE'] = pd.to_numeric(economic_data['UNRATE'],errors='coerce')
economic_data['DCOILBRENTEU'] = pd.to_numeric(economic_data['DCOILBRENTEU'],errors='coerce')
economic_data['T10Y2Y'] = pd.to_numeric(economic_data['T10Y2Y'],errors='coerce')
economic_data['VIXCLS'] = pd.to_numeric(economic_data['VIXCLS'],errors='coerce')
#print(combined_data1)

# Augmented Dickey-Fuller Test on Economic Time Series
def adf_test(series, title=''):
    print(f'ADF Test on {title}')
    result = adfuller(series.dropna(), autolag='AIC')  # Drop missing values
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    if result[1] > 0.05:
        print(f'{title} is not stationary. Consider differencing.\n')
    else:
        print(f'{title} is stationary.\n')

# Running ADF test on each economic series
adf_test(economic_data['UNRATE'].dropna(), 'UNRATE')
adf_test(economic_data['DCOILBRENTEU'].dropna(), 'DCOILBRENTEU')
adf_test(economic_data['GDP_Growth'].dropna(), 'GDP Growth')
adf_test(economic_data['T10Y2Y'].dropna(), 'T10Y2Y')
adf_test(economic_data['VIXCLS'].dropna(), 'VIXCLS')

#take 1st differences
economic_data['UNRATE_diff'] = economic_data['UNRATE'].diff().dropna()
economic_data['DCOILBRENTEU_diff'] = economic_data['DCOILBRENTEU'].diff().dropna()
economic_data['T10Y2Y_diff'] = economic_data['T10Y2Y'].diff().dropna()
economic_data['VIXCLS_diff'] = economic_data['VIXCLS'].diff().dropna()

# Running ADF test on each differenced series
print("ADF Test on Differenced Series")
adf_test(economic_data['UNRATE_diff'], 'UNRATE_diff')
adf_test(economic_data['DCOILBRENTEU_diff'], 'DCOILBRENTEU_diff')
adf_test(economic_data['T10Y2Y_diff'], 'T10Y2Y_diff')
adf_test(economic_data['VIXCLS_diff'], 'VIXCLS_diff')
```

```python
# Test data1
data1['DATE'] = pd.to_datetime(data1['DATE'])
combined_data1 = pd.merge(data1, economic_data, on='DATE', how='inner')

import statsmodels.api as sm
from itertools import combinations

# Assuming 'charge_off_percentage' is the target variable
# And the first difference of non-stationary series have been appended to combined_data1

# Prepare predictors list (excluding 'DATE' and the target variable)
predictors = ['UNRATE_diff', 'DCOILBRENTEU_diff', 'T10Y2Y_diff', 'VIXCLS_diff', 'GDP_Growth']

# Add a lagged version of the target variable as a predictor for AR1 model
combined_data1['lagged_charge_off'] = combined_data1['charge_off_percentage'].shift(1)

# Initialize a placeholder for model comparison
best_model = None
best_r_squared = -1
# Initialize variables to store the best model's details
best_model_predictors = []

# Generate all combinations of three predictors
for combo in combinations(predictors, 3):
    # Ensure the dependent variable is not NaN due to lagging
    df_model = combined_data1.dropna(subset=['lagged_charge_off', 'charge_off_percentage'] + list(combo))
    
    # Define the model
    X = df_model[['lagged_charge_off'] + list(combo)]
    X = sm.add_constant(X)  # Adds a constant term to the predictor
    y = df_model['charge_off_percentage']
    
    model = sm.OLS(y, X).fit()
    
    # Optionally print out each model's summary
    # print(model.summary())
    
# Inside the loop, after finding a better model, update these variables
    if model.rsquared > best_r_squared:
        best_r_squared = model.rsquared
        best_model = model
        best_model_predictors = list(combo)  # Store the predictors of the best model

# After finding the best model, print its summary and the predictors
print("Best Model Summary:")
print(best_model.summary())
print("\nBest 3-Factor Combination:")
print(best_model_predictors)
```
